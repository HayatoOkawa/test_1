```python
def dereat_ward(str_1):
    str_1 = unicodedata.normalize("NFKC", str_1)
    str_1 = str_1.replace('&lt;', '')
    str_1 = str_1.replace('&gt;', '')
    str_1 = str_1.replace('【', '')
    str_1 = str_1.replace('】', '')
    #str_1 = str_1.replace('(', '')
    #str_1 = str_1.replace(')', '')
    str_1 = str_1.replace('<', '')
    str_1 = str_1.replace('>', '')
    str_1 = str_1.replace('.', '')
    #str_1 = str_1.replace('のみ', '')
    #str_1 = str_1.replace('入り', '')
    #str_1 = str_1.replace('入', '')
    #str_1 = str_1.replace('便', '')
    #str_1 = str_1.replace('+', '')
    str_1 = str_1.replace('[', '')
    str_1 = str_1.replace(']', '')
    str_1 = str_1.replace('img src="/resource/cancel" alt=" " border="0"/', '')
    str_1 = str_1.replace('&lt;', '')
    str_1 = str_1.replace('&gt;', '')
    return str_1
```
